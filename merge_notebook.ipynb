```python
import pandas as pd

dataframe1 = pd.read_csv("li-ion_batteries.csv")
dataframe1.head(10)
```

```python
dataframe2 = pd.read_csv("li-ion_batteries_cycling_stats_aggregated.csv")
dataframe2.head(10)
```

```python
dataframe1.shape
```

```python
dataframe2.shape
```

```python
result_dataframe = pd.merge(dataframe1,dataframe2, on="Cell ID")
```

```python
result_dataframe.shape
```

```python
result_dataframe.tail(150)
```

```python
result_dataframe = result_dataframe.rename(columns={'Temperature (C)':'Temperature', 'Charge Rate (C)':'Charge_rate', 'Discharge Rate (C)':'Discharge_rate'})
```

```python
result_dataframe = result_dataframe.rename(columns={'Cell ID':'Cell_ID', 'Max SOC':'Max_SOC', 'Min SOC':'Min_SOC', 'Form Factor':'Form_factor'})
```

```python
result_dataframe = result_dataframe.drop('Unnamed: 0', axis=1)
```

```python
result_dataframe['Cell_model'] = result_dataframe['Cell_ID'].str[:-2]
result_dataframe.head(5)
```

```python
cols = result_dataframe.columns.tolist()
cols
```

```python
cols.remove('Internal_index')
cols.insert(0, 'Internal_index')
cols.remove('Cell_model')
cols.insert(2, 'Cell_model')
cols
```

```python
result_dataframe = result_dataframe[cols]
result_dataframe.head(5)
```

```python
result_dataframe.to_csv('li-ion_batteries_cycling_stats+metadata_aggregated.csv', index = False)
```
