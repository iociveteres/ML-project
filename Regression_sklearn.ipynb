```python
# Necessary imports
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the data
dataset = pd.read_csv("li-ion_batteries_cycling_stats+metadata_aggregated.csv")
dataset.head(10)
```

```python
dataset.dtypes
```

```python
dataset.drop(columns=['Internal_index', 'Cell_ID', 'Cell_model', 'Source'],inplace=True)
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset.head()
```

```python
c=dataset.corr()
plt.figure(figsize = (25,25))
sns.heatmap(c, annot=True)
```

```python
print(dataset.corr()['Charge_capacity_ratio'].sort_values (ascending=False))
```

```python
dataset.drop(columns=['Discharge_capacity_ratio',
                      'Discharge_energy_ratio',
                      'Charge_energy_ratio',
                      'Charge_capacity',
                      'Discharge_capacity',
                      'Discharge_energy',
                      'Charge_energy',
                      'Efficiency',
                      'Max_current',
                      'Min_SOC',
                      'Max_SOC',
                      'Cycle_time',], inplace=True)
```

```python
from sklearn import metrics
def show_metrics(labels_true, labels_pred):
    print("Mean absolute error: " + str(metrics.mean_absolute_error(labels_true, labels_pred)), end='\n')
    print("Mean squared error: " + str(metrics.mean_squared_error(labels_true, labels_pred)), end='\n')
    print("Mean absolute percentage error: " + str(metrics.mean_absolute_percentage_error(labels_true, labels_pred)), end='\n')
    print("Median absolute error: " + str(metrics.median_absolute_error(labels_true, labels_pred)), end='\n')
    print("r2 score: " + str(metrics.r2_score(labels_true, labels_pred)), end='\n')
```

```python
X = dataset.copy()
y = X.pop("Charge_capacity_ratio")
```

```python
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr = LinearRegression()
lr_params = {
}

lr_grid = GridSearchCV(lr, lr_params, n_jobs=-1, cv = 5)
lr_grid.fit(X_train, y_train)
lr_best = lr_grid.best_estimator_
lr_best.get_params()

lr_predicted = lr_best.predict(X_test)
lr_score = lr_best.score(X_test, y_test)
show_metrics(y_test, lr_predicted)
```

```python
elastic_net = ElasticNet()
elastic_net_params = {
    "l1_ratio": [0.3, 0.5, 0.7],
    'alpha': [0.01, 0.1, 0.5, 1, 2],
}

elastic_net_grid = GridSearchCV(elastic_net, elastic_net_params, n_jobs=-1, cv = 5)
elastic_net_grid.fit(X_train, y_train)
elastic_net_best = elastic_net_grid.best_estimator_
elastic_net_best.get_params()

elastic_net_predicted = elastic_net_best.predict(X_test)
elastic_net_score = elastic_net_best.score(X_test, y_test)
show_metrics(y_test, elastic_net_predicted)
elastic_net_best.get_params()
```

```python
ridge = Ridge()
ridge_params = {
    'alpha': [0.01, 0.1, 0.5, 1, 2],
    'solver' : ['svd', 'cholesky', 'sag', 'saga']
}

ridge_grid = GridSearchCV(ridge, ridge_params, n_jobs=-1, cv = 5)
ridge_grid.fit(X_train, y_train)
ridge_best = ridge_grid.best_estimator_
ridge_best.get_params()

ridge_predicted = ridge_best.predict(X_test)
ridge_score = ridge_best.score(X_test, y_test)
show_metrics(y_test, ridge_predicted)
print(ridge_best.get_params())
```

```python
lasso = Lasso()
lasso_params = {
    'alpha': [0.01, 0.1, 0.5, 1, 2],
    'max_iter': [500, 1000, 2000, 3000]
}

lasso_grid = GridSearchCV(lasso, lasso_params, n_jobs=-1, cv = 5)
lasso_grid.fit(X_train, y_train)
lasso_best = lasso_grid.best_estimator_
lasso_best.get_params()

lasso_predicted = lasso_best.predict(X_test)
lasso_score = lasso_best.score(X_test, y_test)
show_metrics(y_test, lasso_predicted)
print(lasso_best.get_params())
```

```python
dt = DecisionTreeRegressor()
dt_params = {
    'criterion': ['squared_error' , 'friedman_mse', 'absolute_error'],
    'max_depth': [30, 40, 50, 65, 80],
    'min_samples_split': range(2, 6, 2),
    'min_samples_leaf': range(1, 5, 2)
}

dt_grid = GridSearchCV(dt, dt_params, n_jobs=-1, cv = 5)
dt_grid.fit(X_train, y_train)
dt_best = dt_grid.best_estimator_
dt_best.get_params()

dt_predicted = dt_best.predict(X_test)
dt_score = dt_best.score(X_test, y_test)
show_metrics(y_test, dt_predicted)
print(dt_best.get_params())
```

```python
xgboost_params = {
    "n_estimators": [100, 150, 200],
    "max_depth": [10, 15],
    "min_samples_split": [2, 5],
    "learning_rate": [0.01, 0.1],
    "loss": ["squared_error"],
    "max_features": ['auto', 'sqrt', 'log2']
}

gsc = GridSearchCV(
    estimator = GradientBoostingRegressor(),
    param_grid = xgboost_params,
    cv=4, n_jobs=-1)


fit=gsc.fit(X_train, y_train)

pred = gsc.predict(X_test)

show_metrics(y_test, pred)
print(fit.best_params_)
```
