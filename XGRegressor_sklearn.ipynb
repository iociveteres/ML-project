```python
# Necessary imports
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the data
dataset = pd.read_csv("li-ion_batteries_cycling_stats+metadata_aggregated.csv")
dataset.head(10)
```

```python
dataset.dtypes
```

```python
dataset.drop(columns=['Internal_index', 'Cell_ID', 'Cell_model', 'Source'],inplace=True)
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset.head()
```

```python
c=dataset.corr()
plt.figure(figsize = (25,25))
sns.heatmap(c, annot=True)
```

```python
print(dataset.corr()['Charge_capacity_ratio'].sort_values (ascending=False))
```

```python
dataset.drop(columns=['Discharge_capacity_ratio',
                      'Discharge_energy_ratio',
                      'Charge_energy_ratio',
                      'Max_SOC',
                      'Charge_capacity',
                      'Discharge_capacity',
                      'Discharge_energy',
                      'Charge_energy',
                      'Charge_rate',
                      'Temperature',
                      'Discharge_rate',
                      'Cycle',
                      'Min_SOC',
                      'Min_voltage',
                      'Cathode_NCA','Form_factor_pouch','Cathode_NMC-LCO'], inplace=True)
```

```python
from sklearn import metrics
def show_metrics(labels_true, labels_pred):
    print("Mean absolute error: " + str(metrics.mean_absolute_error(labels_true, labels_pred)), end='\n')
    print("Mean squared error: " + str(metrics.mean_squared_error(labels_true, labels_pred)), end='\n')
    print("Mean absolute percentage error: " + str(metrics.mean_absolute_percentage_error(labels_true, labels_pred)), end='\n')
    print("Median absolute error: " + str(metrics.median_absolute_error(labels_true, labels_pred)), end='\n')
    print("r2 score: " + str(metrics.r2_score(labels_true, labels_pred)), end='\n')
```

```python
X = dataset.copy()
y = X.pop("Charge_capacity_ratio")
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
train_X, test_X, train_y, test_y = train_test_split(X, y,
                                                    test_size = 0.3, random_state = 123)
xgboost_params = {
    "n_estimators": [50, 100, 150, 200],
    "max_depth": [4, 10, 15],
    "min_samples_split": [2, 5],
    "learning_rate": [0.01, 0.1],
    "loss": ["squared_error"],
    "max_features": ['auto', 'sqrt', 'log2']
}

gsc = GridSearchCV(
    estimator= GradientBoostingRegressor(),
    param_grid=xgboost_params,
    cv=3, n_jobs=-1)


fit=gsc.fit(train_X, train_y)

pred = gsc.predict(test_X)

show_metrics(test_y, pred)
```
